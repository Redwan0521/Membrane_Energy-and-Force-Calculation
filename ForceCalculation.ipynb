```python
import igl
import scipy as sp
import numpy as np
from meshplot import plot, subplot, interact
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
import os
root_folder = os.getcwd()
```

```python
## Load a mesh in OFF format
v, f = igl.read_triangle_mesh(os.path.join(root_folder, "BigSphere.off"))

## Print the vertices and faces matrices 
print("Vertices: ", len(v))
print("Faces: ", len(f))
```

```python
v, f = igl.read_triangle_mesh(os.path.join(root_folder, "BigSphere.off"))
####Gaussian Curvature Calculation___ K= k1*k2
k = igl.gaussian_curvature(v, f)
#plot(v, f, k)



###3 Mean_Curv

l = igl.cotmatrix(v, f) ###laplacian-operator
m = igl.massmatrix(v, f, igl.MASSMATRIX_TYPE_VORONOI)

minv = sp.sparse.diags(1 / m.diagonal())

hn = -minv.dot(l.dot(v))
### Mean Curvature Calculation
h_mean = np.linalg.norm(hn, axis=1)

###Voronoi Area Calculation
area_voronoi=m.diagonal()

type(area_voronoi)
```

```python
### Calculation of Force... Guckenberger et al. 2016 .....Eqn(10)
Lap_H=minv*l*(h_mean/2)
len(Lap_H)
first_product=h_mean*(((h_mean/2)**2)-k)
second_product=Lap_H+first_product


n = igl.per_vertex_normals(v, f)
normal_v=np.linalg.norm(n,axis=1)
type(normal_v)
```

```python
### Force Density and Nodal Force
Kb=1
Force_density=[] # Force_Density
Force_Nodal=[]

for i in range (len(v)):
    Force_density.append((2*Kb)*second_product[i]*normal_v[i])
    Force_Nodal.append(Force_density[i]*area_voronoi[i])
    #print(Force_density[i]) 
    print(Force_Nodal[i])    
```

```python
np.savetxt('Force_txt_finer.txt', Force_density, delimiter=',')
```
